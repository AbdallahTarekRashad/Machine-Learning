#### read_from_kaggle
to rescale image to gray scale and resize to (50,50)

```python
def read_from_kaggle(dir):
    dirList = os.listdir(dir)
    data = list()
    for file in dirList:
        image = cv.imread(dir+"/"+file)
        image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
        image = cv.resize(image,(50,50), interpolation = cv.INTER_AREA)
        data.append(image)
    data = np.array(data)
    return data 
```

```python
  
  def ANN(self,input_shape, n_classes,_lambda):
        self.model = Sequential()
        self.model.add(Dense(512, activation='relu',activity_regularizer=regularizers.l1(_lambda),kernel_regularizer=regularizers.l2(_lambda), input_shape=(input_shape,)))
        self.model.add(Dropout(0.5))
        self.model.add(Dense(256,activity_regularizer=regularizers.l1(_lambda),kernel_regularizer=regularizers.l2(_lambda), activation='relu'))
        self.model.add(Dropout(0.25))
        self.model.add(Dense(n_classes,activity_regularizer=regularizers.l1(_lambda), kernel_regularizer=regularizers.l2(_lambda),activation='softmax'))
        self.model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
        
    def CNN_2D(self,input_shape, n_classes,_lambda):
        self.model = Sequential()
        self.model.add(Conv2D(filters = 32,kernel_regularizer=regularizers.l2(_lambda), kernel_size = (5,5),padding = 'Same',activation ='relu', input_shape = (150,150,3)))
        self.model.add(MaxPooling2D(pool_size=(2,2)))

        self.model.add(Conv2D(filters = 64, kernel_regularizer=regularizers.l2(_lambda),kernel_size = (3,3),padding = 'Same',activation ='relu'))
        self.model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

        self.model.add(Flatten())
        self.model.add(Dense(512,activity_regularizer=regularizers.l2(_lambda)))
        self.model.add(Activation('relu'))
        self.model.add(Dense(5,activity_regularizer=regularizers.l2(_lambda), activation = "softmax"))
        self.model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    def CNN_3D(self,input_shape, n_classes,_lambda):
        input_layer = Input(input_shape)
        
        conv_layer1 = Conv3D(filters=8, kernel_size=(3, 3, 3), activation='relu')(input_layer)
        conv_layer2 = Conv3D(filters=16, kernel_size=(3, 3, 3), activation='relu')(conv_layer1)
        
        pooling_layer1 = MaxPool3D(pool_size=(2, 2, 2))(conv_layer2)
        
        conv_layer3 = Conv3D(filters=32, kernel_size=(3, 3, 3), activation='relu')(pooling_layer1)
        conv_layer4 = Conv3D(filters=64, kernel_size=(3, 3, 3), activation='relu')(conv_layer3)
        
        pooling_layer2 = MaxPool3D(pool_size=(2, 2, 2))(conv_layer4) 
        pooling_layer2 = BatchNormalization()(pooling_layer2)
        
        flatten_layer = Flatten()(pooling_layer2)
        
        dense_layer1 = Dense(units=2048, activation='relu')(flatten_layer)
        
        dense_layer1 = Dropout(0.4)(dense_layer1)
        
        dense_layer2 = Dense(units=512, activation='relu')(dense_layer1)
        
        dense_layer2 = Dropout(0.4)(dense_layer2)
        
        output_layer = Dense(n_classes, activation='softmax')(dense_layer2)
        
        self.model = Model(inputs=input_layer, outputs=output_layer)
        
        self.model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
        
    def logistic_regression():
        pass
    def naive_bayes(self, X, Y):
        pass
    def visualize():
        pass
```

#### Our class

```python
class model():
    allBias = list()
    allVar = list()
    allErr = list()
    
    def __init__(self,n_epochs,batch_size,model_type='SVM'):
        self.n_epochs = n_epochs
        self.batch_size = batch_size
        self.model_type=model_type
    
    def read_data(self,x,y):
        self.tow_d_shape = x.shape[1:]
        self.one_d_shape = x.shape[1]*x.shape[2]
        self.n_classes =np.unique(y).shape[0]
        self.x = x
        self.y = y
        
    def SVM(self,  _lambda=.001):
        # Create the model(SVM)
        self.model = Sequential()
        self.model.add(Dense(self.n_classes, activation='softmax', kernel_regularizer=regularizers.l1(_lambda), input_shape=(self.one_d_shape,)))
        self.model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
        
    
    def fit(self,x_train,y_train):
        if self.model_type == 'SVM':
          x_train = x_train.reshape(x_train.shape[0],self.one_d_shape)
        self.model.fit(x_train, y_train, batch_size=self.batch_size, nb_epoch=self.n_epochs, verbose=1)

    def evaluate(self,x_test,y_test):
        if self.model_type == 'SVM':
          x_test = x_test.reshape(x_test.shape[0],self.one_d_shape)
        scoer = self.model.evaluate(x_test, y_test, verbose=0)
        return scoer

    def bias(self,acc):
        acc = np.array(acc)
        return acc.mean()

    def variance(self,acc):
        acc = np.array(acc)
        return acc.var()
      
    def error(self,err):
        err = np.array(err)
        return err.mean()
      
    def cross_validation(self,n_splits,x,y,_lambda=.001):
        acc = list()
        err = list()
        kfold = StratifiedKFold(n_splits=n_splits, shuffle=True)
        for train, test in kfold.split(x, y):
            
            self.SVM(_lambda)
            
            y_train = np_utils.to_categorical(y[train], self.n_classes)
            
            self.fit(x[train],y_train)
            
            y_test = np_utils.to_categorical(y[test], self.n_classes)
            
            ev = self.evaluate(x[test],y_test)
            
            acc.append(ev[1])
            err.append(ev[0])
            
        self.allBias.append(self.bias(acc))
        self.allVar.append(self.variance(acc))
        self.allErr.append(self.error(err))
        
    def regularizer(self, _range, kFolds):
        for l in _range:
            print('*****************************************************************************')
            print(l)
            self.cross_validation(kFolds,self.x, self.y,l)

    def subset_selection(self):
        for j in range(1, self.tow_d_shape[0]+1, 1):
            input_x = self.x[:,:j,:j]
            self.one_d_shape = input_x.shape[1]*input_x.shape[1]
            
            x_train,x_test,y_train,y_test = train_test_split(input_x, self.y, test_size=0.33)
            self.SVM()
            
            y_train = np_utils.to_categorical(y_train, self.n_classes)
            
            self.fit(x_train,y_train)
            
            y_test = np_utils.to_categorical(y_test, self.n_classes)
            
            ev = self.evaluate(x_test,y_test)
            
            self.allErr.append(ev[0])
            
    def smooth(self, y, box_pts):
        box = np.ones(box_pts) / box_pts
        y_smooth = np.convolve(y, box, mode='same')
        return y_smooth

    def plot(self, style = 'default', smoothing = 50):
        biases = np.array(self.allBias)
        variances = np.array(self.allVar)
        errors = np.array(self.allErr)
        biases = (biases - biases.mean()) / biases.std()
        errors = (errors - errors.mean()) / errors.std()
        variances = (variances - variances.mean()) / variances.std()
        mpl.style.use(style)
        fig, ax = plt.subplots(figsize=(3, 3))
        ax.set_title('Bias Variance Decomposition', color='C0')
        ax.plot(self.smooth(biases, smoothing), label='Bias')
        ax.plot(self.smooth(errors, smoothing), label='Error')
        ax.plot(self.smooth(variances, smoothing), label='Variance')
        ax.legend()
```
